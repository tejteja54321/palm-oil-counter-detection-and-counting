```python
from ultralytics import YOLO

# Load YOLOv10n model
model = YOLO('yolov10n.pt')  # Replace with your YOLOv10n weights file

```

```python
import cv2
import torch
from ultralytics import YOLO
import numpy as np

# Load the trained model
model = YOLO('F:/TEJA/IP/PROJ_10/runs/detect/train3/weights/best.pt')  # Path to trained weights

# Load the video
video_path = 'F:/TEJA/IP/PROJ_10/static/images/img/p3.mp4'  # Path to your video file
cap = cv2.VideoCapture(video_path)

# Check if the video is opened correctly
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Initialize a dictionary to keep track of the class counts
class_counts = {}

# List to store unique objects (bounding boxes) detected
detected_objects = []

# Loop over video frames
while True:
    ret, frame = cap.read()  # Read the next frame
    if not ret:
        break  # If no frame is returned, exit the loop (end of video)

    # Run inference on the frame
    results = model(frame)

    # Extract detections (bounding boxes and class IDs)
    detections = results[0].boxes.data.cpu().numpy()

    # Count objects per class for the current frame
    current_frame_class_counts = {}

    # Loop through each detection
    for detection in detections:
        class_id = int(detection[5])  # Class ID
        x1, y1, x2, y2 = detection[:4]  # Bounding box coordinates

        # Check if the detected object is already counted (based on bounding box)
        object_detected = False
        for obj in detected_objects:
            # If the object is detected and its bounding box overlaps, it's already counted
            if (x1 > obj[0] and y1 > obj[1] and x2 < obj[2] and y2 < obj[3]):  # Simplified overlap check
                object_detected = True
                break

        # If not detected previously, count it as a new object
        if not object_detected:
            detected_objects.append([x1, y1, x2, y2])  # Save the bounding box for future comparison
            current_frame_class_counts[class_id] = current_frame_class_counts.get(class_id, 0) + 1

    # Update the cumulative class counts
    for class_id, count in current_frame_class_counts.items():
        class_counts[class_id] = class_counts.get(class_id, 0) + count

    # Display the cumulative counts on the frame
    for class_id, count in class_counts.items():
        cv2.putText(frame, f"Class {class_id}: {count}", (10, 30 + class_id * 30), 
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Visualize results on the frame
    annotated_frame = results[0].plot()  # Annotated frame with boxes and labels

    # Display the frame
    cv2.imshow("Detections", annotated_frame)

    # Press 'q' to quit the video display
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
